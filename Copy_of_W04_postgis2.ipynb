1. **What subway station is in \'Little Italy\'? What subway route is it on?**

```python
#essentially join subway stations with boroughs based on geometry
#select name and routes from subway stations
#join based on st contains geometries from n and s tables
#where the neighbiourhood name is littel italy

con.sql("""
SELECT s.name, s.routes
FROM nyc_subway_stations AS s
JOIN nyc_neighborhoods AS n
ON ST_Contains(n.geom, s.geom)
WHERE n.name = 'Little Italy';
""")
```

2. **What are all the neighborhoods served by the 6-train?** (Hint: The `routes` column in the `nyc_subway_stations` table has values like \'B,D,6,V\' and \'C,6\')


```python
#The routes column in the nyc_subway_stations table has values like 'B,D,6,V' and 'C,6')
#select routes contain 6 string from nyc subway stations
#join subway stations to neighbourhoods
#where routes 6


con.sql("""
SELECT DISTINCT n.name
FROM nyc_neighborhoods AS n
JOIN nyc_subway_stations AS s
  ON ST_Intersects(n.geom, s.geom)
WHERE s.routes LIKE '%6%';
""")
```

3. **After 9/11, the \'Battery Park\' neighborhood was off limits for several days. How many people had to be evacuated?**

```python
#how many people in battery park - so table for neighbourhoods and table for census blocks
#census blocs is polygon  so is neighbourhood


con.sql("""
SELECT SUM(p.popn_total),
FROM nyc_census_blocks AS p
JOIN nyc_neighborhoods AS n
  ON ST_Intersects(n.geom, p.geom)
WHERE n.name = 'Battery Park';
""")
```

4. **What neighborhood has the highest population density (persons/km2)?**


```python
#join census blocks to neighbourhoods
#sum pop total by neighbourhood and divide by area of neighbourhood (st_area km2 /1000000)

con.sql("""
SELECT n.name, SUM(p.popn_total)/ST_Area(n.geom)/1000000.0 AS popKM2
FROM nyc_census_blocks AS p
JOIN nyc_neighborhoods AS n
  ON ST_Intersects(p.geom, n.geom)
GROUP BY n.name, n.geom
ORDER BY popKM2 DESC LIMIT 2;
""")
```
